```python
# Log formatını düzenleyecek regex kalıbı
log_pattern = re.compile(r'(\S+) (\S+) (\S+) \[(.*?)\] "(.*?)" (\d{3}) (\S+)')
```

```python
# Mevcut 'log' sütununu regex ile parçalara ayırarak yeni sütunlar oluşturma
df[['ip', 'identity', 'user', 'timestamp', 'request', 'status_code', 'size']] = df['log'].str.extract(log_pattern)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
# Sentence-BERT modelini yükleyelim
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
# Request sütunundaki verileri vektörleştirme
request_embeddings = model.encode(df['request'].tolist())
```

```python
# Vektörlerin boyutunu kontrol edelim
print(request_embeddings.shape)
```

```python
!pip install faiss-cpu
```

```python
import faiss
```

```python
# FAISS index'i oluşturma (L2 distance için)
index = faiss.IndexFlatL2(request_embeddings.shape[1])
```

```python
# Vektörleri FAISS index'ine ekleme
index.add(np.array(request_embeddings))
```

```python
# FAISS index'inde kaç vektör olduğunu kontrol etme
print(index.ntotal)
```

```python
# Kullanıcı sorusunu vektörleştirme
user_query = "Which pages were accessed the most?"
query_embedding = model.encode([user_query])
```

```python
# Vektörün boyutunu kontrol edelim
print(query_embedding.shape)
```

```python
# En yakın 5 vektörü bulma
D, I = index.search(query_embedding, k=5)
```

```python
# I çıktısı, en yakın log kayıtlarının indekslerini gösterir
print(I)
```

```python
# Bulunan log kayıtlarını gösterme
similar_logs = df.iloc[I[0]]
print(similar_logs)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
```

```python
# T5 modelini ve tokenizer'ını yükleyelim
model_name = "t5-small"
t5_model = T5ForConditionalGeneration.from_pretrained(model_name)
t5_tokenizer = T5Tokenizer.from_pretrained(model_name)
```

```python
# FAISS'ten aldığımız log kayıtlarını metin formatına çevirelim
log_text = " ".join(similar_logs['request'].tolist())
```

```python
# Sorgu ile birlikte modelin anlaması için bir bağlam oluşturalım
input_text = f"question: Which pages were accessed the most? context: {log_text}"
```

```python
# Girdi metnini tokenize edelim
input_ids = t5_tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
```

```python
# Modelden yanıt alalım
outputs = t5_model.generate(input_ids, max_length=50, num_beams=4, early_stopping=True)
```

```python
# Yanıtı decode edelim
answer = t5_tokenizer.decode(outputs[0], skip_special_tokens=True)
print(answer)
```
